```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"
```

```python
link='https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
table_raw = pd.read_html(link, match='Коефіцієнт народжуваності в', decimal=',', thousands='.')
print(table_raw[0])
```

```python
#type(table_raw)
y_data=table_raw[0].loc[27][1:]
x_data=(table_raw[0].loc[27].index[1:])
plt.plot(
    x_data, y_data, label='% народжуваності', marker='*',
    linestyle='--' )
plt.xlabel('Роки')
plt.ylabel('Коефіцієнт народжуваності в Україні')
plt.title('Коефіцієнт народжуваності в Україні 1950-2019')
plt.grid()
plt.legend()
plt.show()
```

```python
#Вивести перші рядки таблиці за допомогою методу head
print(table_raw[0].head())
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
print(table_raw[0].shape)
```

```python
#Замініть у таблиці значення "—" на значення NaN
table_raw[0].replace(to_replace='—', value=np.nan)
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(table_raw[0].dtypes)
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
table=table_raw[0].replace(to_replace='—', value=np.nan)
table=table.astype({'2014':'float64','2019':'float64'})
table.index=[
'Крим',
'Вінницька',
'Волинська',
'Дніпропетровська',
'Донецька',
'Житомирська',
'Закарпатська',
'Запорізька',
'Івано-Франківська',
'Київська',
'Кіровоградська',
'Луганська',
'Львівська',
'Миколаївська',
'Одеська',
'Полтавська',
'Рівненська',
'Сумська',
'Тернопільська',
'Харківська',
'Херсонська',
'Хмельницька',
'Черкаська',
'Чернівецька',
'Чернігівська',
'Київ',
'Севастополь',
'Україна']
table=table.drop(["Регіон"], axis=1)
print(table)
print(table.dtypes)
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
#Option1
column_index=table.columns
for cl in column_index:
  print(f'частка пропусків міститься в колонці {cl} ', (table[cl].isnull()).sum()/len(table[cl]))  
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
table=table.drop(index='Україна')
print(table)
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
table=table.fillna(table.mean())
print(table)
```

```python
plt.title('Коефіцієнт народжуваності в Україні по регіонам за 2019')
sns.scatterplot(data=table, x='2019', y=table.index)
```

```python
sns.set_style("whitegrid")
plt.title('Коефіцієнт народжуваності в Київі в 1950-2019')
sns.lineplot(x=table.columns, y=table.loc['Київ'], data=table.loc['Київ'])
```

```python
plt.figure(figsize=(5,8))
plt.barh(
    table.index, table['2019'],
    )
plt.grid()
plt. title('Коефіціент народжуваності в регіонах України за 2019')
plt.xlabel('Коефіціент народжуваності')
plt.ylabel('Регіони Україні')
plt.show()
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
print('Середній рівень народжуваності в Україні за 2019', table['2019'].mean())
table['2019'][table['2019']>table['2019'].mean()]
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
print(table['2014'],'\n')
print('Регіон',table['2014'][table['2014']==table['2014'].max()])
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.title('Коефіцієнт народжуваності в Україні по регіонам за 2019')
table['2019'].plot(kind='bar')
```
